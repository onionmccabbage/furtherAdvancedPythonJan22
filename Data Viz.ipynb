# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# for notebooks we need...
%matplotlib inline
```

```python
df_WTI = pd.read_csv('WTI2017.csv', index_col='Date', 
                     parse_dates=True)
```

```python
x = df_WTI.index
y = df_WTI['Price']
df_WTI
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
# generate a chart
plt.plot(x,y, 'r')# r is red
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('2017 closing Prices')
plt.show()
```

```python
plt.subplot(1,2,1)
x = df_WTI.index
y1 = df_WTI['High']
y2 = df_WTI['Low']
plt.plot(x, y1, 'b--')
plt.subplot(1,2,2)
plt.plot(x, y2, 'g*')
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.set_title("WTI - 2017 Closing Price");

axes.set_xlabel("Date")
axes.set_ylabel("Price ($)");

x = df_WTI.index
y1 = df_WTI['High']
y2 = df_WTI['Low']

axes.plot(x, y1, label="High")
axes.plot(x, y2, label="Low")

axes.legend()
```
